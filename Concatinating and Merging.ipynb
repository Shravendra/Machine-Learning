```python
#Left Merge -> 
# Here it will keep only matched record of left and right dataframe.
# It will keep all the left dataframe values and if there us match in right.
# Dataframe it will keep that matched values.
# if there is no matched it will keep with NAN
pd.merge(df4, df5, on = 'PlaceNumber', how  = 'left') 
```

```python
# It will keep all the right dataframe values and if there us match in left.
# Dataframe it will keep that matched values.
# if there is no matched it will keep with NAN.
pd.merge(df4, df5, on = 'PlaceNumber', how = 'right') 
```

```python
# combiation of left and right merge.
pd.merge(df4, df5, on = 'PlaceNumber', how = 'outer') 
```
