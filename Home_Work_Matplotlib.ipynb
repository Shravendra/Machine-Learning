```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Draw a bar graph for chocolate lover.
chocolate = ['Dairy Milk', 'Kitkat', 'Perk', '5-Star', 'Snikers', 'lolipop']
nos_of_sell = [2000, 1200, 2400, 1500, 1900, 2100]

plt.title("Number of Sells of Chocolate in India per day", fontsize = 20, color = 'g')
plt.xlabel("Name of Chcolate", fontsize = 10, color = 'b')
plt.ylabel("Sells in Per Day", fontsize = 10, color = 'b')

my_col = ['orange', 'red', 'purple', 'brown', 'cyan', 'green']

plt.bar(chocolate, nos_of_sell, color  = my_col )

plt.legend(title = 'Name of Chocolate')
plt.show()
```
