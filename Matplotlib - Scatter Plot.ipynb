```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Age_of_car = [2, 4, 1, 6, 7, 8, 9, 10, 3, 2]
speed = [80, 60, 100, 50, 50, 50, 40, 30, 80, 80]
plt.scatter(Age_of_car, speed)
plt.show()
```

```python
x = np.random.randint(10, 1000, 100)
y = np.random.randint(10, 1000, 100)
size = np.random.randint(1, 200, 100)
plt.figure(figsize = (10, 5))
plt.scatter(x, y, alpha = 0.4, color  = 'r', label = 'x vs y', s = size) # alpha - is used for transparancy.

x1 = np.random.randint(20, 1000, 200)
y1 = np.random.randint(20, 1000, 200)
size1 = np.random.randint(1, 200, 200)
plt.scatter(x1, y1, color = 'g', label = 'x1 vs y1', s = size1)

x2 = np.random.randint(50, 1000, 100)
y2 = np.random.randint(50, 1000, 100)
plt.scatter(x2, y2, color = 'orange', label = 'x2 vs y2')

plt.legend()
plt.show()
```
