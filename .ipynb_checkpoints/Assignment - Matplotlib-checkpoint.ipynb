```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("tips.csv")
df.head()
```

```python
# Check how many unique values present in time column.
# find the total count of dinner and lunch. 
x = df['time'].unique()
print("Unique values present in time column", x)
dinner = df['time'].value_counts()['Dinner']
lunch = df['time'].value_counts()['Lunch']
print("Number of Dinner Counts is ", dinner)
print("Number of Lunch Counts is ", lunch)
```

```python
# Draw a bar chart for this count.
time = ["Dinner" , "Lunch"] 
nos_of_count = [dinner, lunch]
col = ['orange', 'green']
plt.figure() 
plt.bar(time, nos_of_count, color = col)
plt.show()
```

```python
# Draw a pie chart for tips which was given by female and male.
x = df['gender'].unique()
print("Unique values present in time column is", x)

male = df['gender'].value_counts()['Male']
female = df['gender'].value_counts()['Female']
print("Number of male Counts is ", male)
print("Number of female Counts is ", female)

gender = ['Male', 'Female']
nos_of_tips = [male, female]
mycolors = ["red", "hotpink"]
plt.pie(nos_of_tips, labels = gender, colors = mycolors, startangle=90, rotatelabels = 2, autopct = '%2.1f%%')

plt.legend(title = "Gender:")
plt.show()
```

```python
# Draw a bar chart for total bill vs Day.
total_bill = df['total_bill']
day = df['day']

plt.figure() 
plt.bar(day, total_bill)
plt.show()
```

```python
# Draw a Scatter plot for smoker vs tips.
smoker = df['smoker']
tips = df['tip']


plt.figure(figsize = (10, 5))
plt.scatter(smoker, tips, alpha = 0.4, color  = 'r', label = 'smoker vs tip')
plt.show()
```

```python
# Draw a subplot for gender vs size, gender vs smoker, gender vs tips, gender vs time.

Gender = df['gender']
Size = df['size']
Smoker = df['smoker']
Tips = df['tip']
Time = df['time']

x = [23, 45, 123, 56, 89]
y = [1, 2, 3, 4, 5]
z = [20, 12, 34, 56, 78]
plt.figure(figsize = (20, 10))

# Gender Vs Size
plt.subplot(2, 3, 1)
plt.plot(Gender, Size)

# Gender Vs Smoker
plt.subplot(2, 3, 2)
plt.plot(Gender, Smoker)

# Gender Vs Tips
plt.subplot(2, 3, 3)
plt.plot(Gender, Tips)

# Gender Vs Time
plt.subplot(2, 3, 4)
plt.plot(Gender, Time)

plt. show()
```
