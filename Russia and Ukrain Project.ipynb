```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('russia_losses_equipment.csv')
df.head()
```

```python
fill_dict = {
    'special equipment' : 0,
    'mobile SRBM system' : 1
    }

df.fillna(fill_dict)
```

```python
df.describe()
```

```python
df.corr() # Its used to find the corelation b/w two or more columns.
```

```python
x = [i for i in range(1, len(df)+1)] # List Comprehension technique.
y = df['helicopter']
z = df['tank']
plt.plot(x, y, label = 'Helicopter')
plt.plot(x, z, label = 'Tank')
plt.legend()
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, label = 'Helicopter')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, z, label = 'Tank')
plt.legend()
plt.show()
```
