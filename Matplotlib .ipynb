```python
# Matplotlib : Its a low level graph plotting liberary in python that serves as a visualisation utility.
# It is used to See the relationship of difference b/w the columns.
#1. Line Plot - this is basic plot and represented by line.
#2. Bar Plot/Graph - it is used to draw catogorical data nad numerical data.
#3. Histogram - A histogram is a graph showing frequency distributions. It is a graph showing the number of observations within each given interval.
#4. Pie Plot/Chart - 
#5. Box Plot - 
#6. Scatter Plot - It is a diagram where each value in the data set is represented by dot.

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Line Plot Example
x = [3, 4, 5, 7, 8]
y = [0, 1, 5, 3, 6]

plt.plot(x, y)
plt.show()
```

```python
# Plot with the color named "hotpink":

ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, c = 'hotpink')
plt.show()
```

```python
# Generate some random Number for x-axis and y-axis of array.
x = np.random.normal(0, 50, 10)

y = np.random.randint(10, 200, 10)

plt.plot(x, y, '*') # Marker is used to for design.
plt.show()
```

```python
plt.plot(x, y, 'ro-', lw = 4, ms = 10, mfc = 'b')  #lw = line width, # ms = marker strength
plt.xlabel('x-axis', fontsize = 15)
plt.ylabel('Y-axis', fontsize = 15)
plt.title('Line Plot Example', fontsize = 20, color = 'g')
plt.show()
```

```python
# Plot with a 20.5pt wide line:
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

```python
# multiple line plot
x = np.random.randint(1, 40, 10)
y = np.random.randint(1, 50, 10)
z = np.random.randint(1, 60, 10)
plt.plot(x, y, 'o-.', label = 'x vs y', color = 'g')
plt.plot(x, z, '-', label = 'x vs z', color = 'b')
plt.plot(y, z, label = 'y vs z', color  = 'purple')
plt.plot(x, y, z, label = 'x vs y vs z', color = 'red', linestyle = '--', linewidth = 1.5)
plt.legend(title = "Line Color", loc = 2) #loc = location of legend 
plt.show()
```

```python
x = [1, 4, 6, 8, 7]
y = [i**2 for i in x]
z = [i**3 for i in x]
plt.plot(x, y, label = 'Square')
plt.plot(x, z,label = 'Cube')
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend(title = 'Mathmatical constrained')
plt.tick_params(labelsize = 10)
plt.show()
```

```python
# Draw two lines by specifying a plt.plot() function for each line:
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

```python
# Add Grid Lines to a Plot
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()

plt.show()
```

```python
# Specify Which Grid Lines to Display.

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y, 'r')

plt.grid(axis = 'x')

plt.show()
```

```python
# Display grid line for the y-axis:

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'y')

plt.show()
```

```python
# Set Line Properties for the Grid.

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```

```python
# Scatter Plot
# Draw two plots on the same figure:
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.colorbar()

plt.show()
```

```python
# Create random arrays with 100 values for x-points, y-points, colors and sizes:

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

```python
# Bar Plot/Graph

x = np.array(["Sourabh", "Dheeraj", "Priyatam", "Raju", 'Makhan', 'Priyank', 'Mukesh'])
y = np.array([9, 18, 11, 7, 0.2, 21, 13])

plt.title("Salary in Lakhs", fontsize = 20, color = 'g')
plt.xlabel("Name of Persons", fontsize = 10, color = 'b')
plt.ylabel("Salary in Rupees", fontsize = 10, color = 'b')

c = ['orange', 'red', 'purple', 'blue', 'black', 'green', 'yellow']

plt.bar(x,y, color = c)
plt.show()
```

```python
# Draw 4 horizontal bars:

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, color  = 'red')
plt.show()
```

```python
# Thin Bars - Draw 4 very thin bars:
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

c = ['orange', 'red', 'purple', 'blue']
     
plt.bar(x, y, width = 0.1, color = c)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

c = ['orange', 'red', 'purple', 'blue']

plt.barh(x, y, color  = c, height = 0.1)
plt.show()
```

```python
y = np.array([35, 25, 30, 15])
mylabels = ["Mango", "Orange", "Bananas", "Apples"]
plt.pie(y, labels = mylabels,  startangle = 90)
plt.show() 
```

```python
# Pull the "Mango" wedge 0.2 from the center of the pie:
mylabels = ["Mango", "Orange", "Bananas", "Apples"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels,  startangle = 90, explode = myexplode)
plt.show() 
```

```python
# Add a shadow:
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

```python
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show() 
```
