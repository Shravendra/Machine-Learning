```python
import matplotlib.pyplot as plt
```

```python
bike_name = ['NS200', 'xpluse', 'pulser', 'R15', 'FZ']
sales = [20, 30, 20, 50, 40]
mycolors = ["red", "hotpink", "green", "purple", 'orange']
plt.pie(sales, labels = bike_name, colors = mycolors, startangle=90, rotatelabels = 2, autopct = '%2.1f%%')
# 'startangle' will used to start from given angle.
plt.legend(title = "Bike Name:")
plt.show()
```

```python
myexplode = [0, 0.2, 0, 0, 0]
plt.pie(sales, labels = bike_name, labeldistance = 0.5,
        startangle = 90, explode = myexplode,
        rotatelabels = True, shadow = True)
# 'labeldistance' and 'rotatelabels' is used to dispaly the labels properly.
plt.show()
```
