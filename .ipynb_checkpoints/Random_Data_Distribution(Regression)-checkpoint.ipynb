```python
import numpy
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
x = numpy.random.normal(5.0, 1.0, 1000)

y = numpy.random.normal(10.0, 2.0, 1000)

plt.scatter(x, y)
plt.show()
```

```python
# Linear Regression : Lenear regression used the relationship between the data-points.
                     #to draw a straight line through all them.
    
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 22, 23, 9, 6]
y = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]
plt.scatter(x, y)
plt.show()
```

```python
from scipy import stats
```

```python
# Linear Regression : Lenear regression used the relationship between the data-points.
                     #to draw a straight line through all them.slope

# Execute a method that returns somr=e importsnt key values of linear regression. 
slope, intercept, r, p, std_err = stats.linregress(x, y)

# Create a function that uses the slope and intercept value to return a new value.
def myfunc(x):
    return slope * x + intercept 


# Run each value of the array through the function. This will result in a new array with new values for the y-axis.
mymodel = list(map(myfunc, x))

# Draw the originalscatter plot.
plt.scatter(x, y)

# Draw the Line oflinear regression.
plt.plot(x, mymodel)

#Display the diagram.
plt.show()
```

```python
print(r) # 'r' for relationship
```

```python
Speed = myfunc(10)
print(Speed)
```

```python
x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

# NumPy has a method that lets us make a polynomial model.
mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

# Then specify how the line will display, we start at position 1, and endat position 22.
myline = numpy.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```
