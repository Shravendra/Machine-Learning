```python
#dropna or fillna method:
# dropna : It will delete a row which contains null value or NAN value.
# fillna : It will fill the record for empty cell or NAN value.
# >> If column have 'neumerical' value the null containing rows fill with 'mean/median' value of that column.
# >> If it column have 'string' value the null containing rows fill with 'mode' value of that column.
import pandas as pd
```

```python
df = pd.read_csv('demodataset.csv')
df
```

```python
df['Age'].mean()
```

```python
df['Age'].mean()
fillna_dict = {
    'Age' : df['Age'].mean(),
    'Eligible' : df['Eligible'].median(),
    'Score' : df['Score'].mean(),
    'Location' : df['Location'].mode()[0]
}
result = df.fillna(fillna_dict)
result.to_excel('finaldemodataset.xlsx', index = False)
df1 = pd.read_excel('finaldemodataset.xlsx')
df1
```

```python
# also did with using variable by storing the value in it.
x = df['Age'].mean()
y = df['Eligible'].mean()
z = df['Score'].mean()
a = df['Location'].mode()[0]
fillna_dict = {
    'Age' : x,
    'Eligible' : y,
    'Score' : z,
    'Location' : a
}
df.fillna(fillna_dict)
```

```python
df['Location'].mode()
```

```python
# always start with 0th index position when we using 'Mode' to get accurate result.
m = df['Location'].mode()[0]
m
```

```python
df.dropna() #it used to drop the NaN value containing rows.
```

```python
#if thresh value is greater or equal that total value of row it will keep that row.
#If it is not matching it will delete that rows.
#"thresh = 1" will not drop the rows having atleast one value.
df.dropna(thresh = 1) 
```

```python
df.dropna(thresh = 5) #it will drop only that rows whose having atleast 5 values.
```
