```python
import pandas as pd
```

```python
df = pd.read_csv('CPI.csv')
df
```

```python
import matplotlib.pyplot as plt
```

```python
Jurisdiction = df['Jurisdiction']
Year = df['2015']

fig = plt.figure(figsize = (10, 6))
 
# creating the bar plot
plt.bar(Jurisdiction, Year, color ='maroon',
        width = 0.4)
 
plt.xlabel("Jurisdiction")
plt.ylabel("Year")
plt.title("corruption perception index")
plt.show()
```
